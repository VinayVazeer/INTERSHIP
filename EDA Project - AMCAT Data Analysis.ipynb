```python
all_jobs = Hardware_engineer_jobs + Programmer_analyst_jobs + Software_Engineer_jobs + Associate_Engineer_jobs

avg_salary_by_designation = {}

for job in all_jobs:
    job_data = df[df['Designation'].str.lower() == job]
    avg_salary = job_data['Salary'].mean()
    avg_salary_by_designation[job] = avg_salary

avg_salary_series = pd.Series(avg_salary_by_designation)

# Create a bar plot
plt.figure(figsize=(14, 8))
ax = avg_salary_series.sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Salary by Designation')
plt.xlabel('Designation')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)

# Draw a border line at 250,000
ax.axhline(y=250000, color='red', linestyle='--', linewidth=2, label='Border Line: ₹250,000')

plt.legend()
plt.show()
```

```python
from scipy.stats import f_oneway

programming_analyst_salary = programmer_analyst_data["Salary"]
software_engineer_salary = software_engineer_data["Salary"]
hardware_engineer_salary = hardware_engineer_data["Salary"]
associate_engineer_salary = associate_engineer_data["Salary"]

f_statistic, p_value = f_oneway(programming_analyst_salary,software_engineer_salary,hardware_engineer_salary,associate_engineer_salary)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There are significant differences in average salaries.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in average salaries.")
```

# Is there a relationship between gender and specialization? 

Its a Caterogical vs categorical hence we should perform Chi Square test (Dependnecy test)

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Specialization', hue='Gender', data=df)
plt.title('Specialization vs Gender')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
from scipy.stats import chi2_contingency


contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

print("Contingency Table:")
print(contingency_table)

chi2, p, _, _ = chi2_contingency(contingency_table)

print("\nChi-squared statistic:", chi2)
print("P-value:", p)

alpha = 0.05  # Set the significance level
print("\nSignificance level:", alpha)
if p < alpha:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("There is no significant relationship between Gender and Specialization.")
```

A Chi-Square test, a statistical method for examining the association between categorical variables, was conducted on the variables of Gender and Specialization. The test revealed a statistically a 95% of significant relationship between a candidate's gender and their chosen specialization.
